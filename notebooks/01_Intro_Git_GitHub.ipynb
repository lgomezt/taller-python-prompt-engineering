## 1. El problema que resuelve Git

Imaginen que están trabajando en una tesis con su director. Cada vez que mandan una versión por correo, el archivo termina llamándose algo como:

```
tesis_final.docx
tesis_final_v2.docx
tesis_final_v2_CORREGIDA.docx
tesis_final_v2_CORREGIDA_con_comentarios_director.docx
tesis_ESTA_SI_ES_LA_FINAL.docx
```

Esto pasa **todo el tiempo** en proyectos de código y datos. Y es un desastre cuando:

- Quieren volver a una versión anterior (¿cuál era la buena?).
- Dos personas están editando el mismo archivo a la vez.
- Necesitan saber **quién** cambió **qué** y **cuándo**.
- Quieren probar algo arriesgado sin romper lo que ya funciona.

**Un sistema de control de versiones** (SCV) resuelve exactamente esto: guarda la historia completa de un proyecto, permite volver atrás en el tiempo, y hace posible que varias personas trabajen en paralelo sin pisarse.

Git es, hoy, el sistema de control de versiones más usado en el mundo.


## 2. ¿Qué es Git?

**Git** es un programa que se instala en el computador. Funciona **localmente**: crea una "foto" (llamada *commit*) del estado de una carpeta cada vez que ustedes se lo piden, y guarda toda la historia en una subcarpeta oculta llamada `.git`.

Características clave:

- Fue creado por Linus Torvalds en 2005 (el mismo que creó Linux).
- Es **distribuido**: cada copia del repositorio contiene la historia completa.
- Funciona **sin internet**. Ustedes pueden hacer commits, ramas y navegar la historia totalmente offline.
- Es **gratis** y de código abierto.

> **Analogía**: Git es como la función "Control + Z" de Word, pero con esteroides. Cada commit es un punto de guardado al que pueden volver, y pueden tener múltiples "líneas del tiempo" (ramas) simultáneamente.

### Conceptos básicos de Git

| Concepto | Qué es |
|---|---|
| **Repositorio** (*repo*) | Una carpeta con código bajo control de Git |
| **Commit** | Una foto guardada del proyecto en un momento dado |
| **Branch** (*rama*) | Una línea paralela de desarrollo |
| **Merge** (*fusión*) | Combinar dos ramas |
| **Remote** (*remoto*) | Una copia del repositorio que vive en otro lado (ej. GitHub) |
| **Clone** | Descargar un repositorio remoto al computador |
| **Pull** | Traer cambios del remoto al computador local |
| **Push** | Enviar mis commits locales al remoto |

No se preocupen si esto se ve abrumador. En la práctica, el 90% del tiempo van a usar solo **clone**, **pull**, **commit** y **push**.


## 3. ¿Qué es GitHub?

**GitHub** es una empresa (ahora de Microsoft) que ofrece un **servicio web** para alojar repositorios de Git en la nube. Básicamente es *"Dropbox pero para código con Git"*, más una enorme capa social encima.

Cuando ustedes ponen su repositorio en GitHub obtienen:

- Una URL pública (o privada) para su proyecto: `https://github.com/usuario/repo`.
- Una interfaz web para navegar el código, ver diffs, historial, etc.
- Herramientas de colaboración: *issues*, *pull requests*, *code review*, *discussions*.
- Integración con herramientas de CI/CD (correr tests automáticamente).
- Un portafolio público — muchas empresas miran el perfil de GitHub de los candidatos.
- GitHub Pages (sitios estáticos), GitHub Actions (automatización), Copilot, etc.

### Alternativas a GitHub

GitHub no es el único servicio de este tipo. Otros populares:

- **GitLab** — muy usado en empresas (ofrece versión self-hosted gratuita).
- **Bitbucket** — de Atlassian, integrado con Jira.
- **Codeberg** — alternativa sin fines de lucro.

Todos hablan "Git" de fondo, así que los comandos de git son los mismos. Para esta clase usamos GitHub porque es el estándar de facto en la industria y en la academia.


## 4. Git ≠ GitHub

Esta es la confusión más común entre gente que está arrancando. Vamos a dejarlo claro:

| | **Git** | **GitHub** |
|---|---|---|
| **¿Qué es?** | Un programa (software) | Un servicio web (empresa) |
| **¿Dónde vive?** | En tu computador | En la nube (servidores de Microsoft) |
| **¿Necesita internet?** | No | Sí |
| **¿Quién lo hizo?** | Linus Torvalds (2005) | Empresa GitHub Inc. (2008), hoy de Microsoft |
| **¿Es gratis?** | Sí, siempre | Sí para uso básico; planes pagos |
| **¿Se puede usar sin el otro?** | Sí (Git solo, en local) | No (GitHub usa Git internamente) |

### Una analogía útil

- **Git** es como `Microsoft Word`: un programa que corre en tu computador para manipular archivos.
- **GitHub** es como `OneDrive` o `Google Drive`: un servicio en la nube donde guardas esos archivos para compartirlos.

Uno puede usar Word sin OneDrive. Pero para colaborar con otras personas, OneDrive facilita mucho la vida.

---

### ¿Qué pasa cuando combinamos ambos?

El flujo típico es:

```
┌──────────────────┐                       ┌──────────────────┐
│  MI COMPUTADOR   │                       │     GITHUB       │
│                  │                       │                  │
│   [ Git local ]  │   git push / pull     │  [ Repo remoto ] │
│   historia .git  │ <─────────────────>   │  historia .git   │
│                  │                       │                  │
└──────────────────┘                       └──────────────────┘
```

1. Yo trabajo en mi computador y hago commits con Git (local, sin internet).
2. Cuando quiero compartir mis cambios, hago `git push` → los subo a GitHub.
3. Cuando otro colaborador hizo cambios, hago `git pull` → los bajo a mi computador.


## 5. ¿Para qué se usan Git y GitHub en Data Science?

En un proyecto típico de ciencia de datos, git + GitHub se usan para:

1. **Versionar código** — scripts de Python, notebooks, SQL, etc.
2. **Colaborar** — varias personas trabajando en el mismo proyecto sin pisarse.
3. **Reproducibilidad** — cualquier persona (o ustedes mismos 6 meses después) puede clonar el repo y correr el análisis.
4. **Code review** — un colega revisa tu código antes de que se integre al proyecto principal.
5. **Portafolio** — sus proyectos en GitHub son el currículum técnico que los reclutadores miran.
6. **Integración con AI tools** — **Claude Code, Cursor, Antigravity y Copilot funcionan sobre git**. Son herramientas que *entienden* la historia de commits, las ramas, etc. No usar git significa no aprovechar la mitad de lo que ofrecen.

### ¿Y los datos?

Aquí hay una regla importante y vale la pena detenerse:


## 6. ⚠️ GitHub NO es un repositorio para compartir datos

Esta es una **regla fundamental** que muchas personas violan al arrancar:

> **GitHub es para código. NO es para datos.**

### ¿Por qué?

1. **Tamaño**. Git fue diseñado para archivos de texto (código). Archivos grandes (>100 MB) simplemente no se pueden subir — GitHub los rechaza. Y versionar un CSV de 5 GB es absurdo: cada commit duplicaría el archivo en la historia y el repo crecería descontroladamente.

2. **Datos sensibles**. Datos personales, médicos, financieros, etc. están sujetos a leyes (Habeas Data en Colombia, GDPR en Europa, HIPAA en EE.UU.). Subirlos a un servidor público o incluso privado puede ser **ilegal**. Una vez pushean algo a GitHub, es casi imposible borrarlo del historial.

3. **Reproducibilidad real**. Si los datos dependen de una consulta SQL, de una API o de un archivo que cambia, versionar una copia estática engaña a los colaboradores. Mejor versionar el **código que obtiene los datos**, no los datos.

4. **Costos y licencias**. Muchos datasets tienen términos de uso que prohíben redistribuirlos.

### ¿Entonces dónde se guardan los datos?

| Tipo | Herramientas típicas |
|---|---|
| Datasets grandes | AWS S3, Google Cloud Storage, Azure Blob, OneDrive empresarial |
| Datos tabulares | Bases de datos (PostgreSQL, BigQuery, Snowflake) |
| Datos versionados | DVC (Data Version Control), Git LFS, LakeFS |
| Datos privados/sensibles | Servidores internos con control de acceso |

El código sí vive en Git/GitHub; los datos viven en otro lado, y el código hace referencia a ellos (ej. `s3://my-bucket/data.csv`).

### ⚠️ Excepción en esta clase

Para **esta sesión específica** vamos a subir los dos CSVs al repositorio porque:

- Son pequeños (<5 MB en total).
- Son públicos y sintéticos (no son datos personales reales).
- Queremos que clonar el repo sea un solo paso, sin configurar credenciales de AWS ni nada externo.

**Esto es una decisión pedagógica, no una buena práctica**. En un proyecto real tendrían el código en GitHub y los datos en otro lado. Tenerlo en mente.

## 8. Paso 2 — Instalar Git

Git es un programa que se corre desde la **terminal** (línea de comandos). Los pasos son distintos para Mac y Windows.

### 8.1 Verificar si ya está instalado

Antes de instalar nada, revisen si ya tienen Git. Abran una terminal y corran:

**Mac**: abran `Terminal.app` (Spotlight → "Terminal").

**Windows**: abran `PowerShell` (tecla Win → escriban "PowerShell").

Y tipeen:

```bash
git --version
```

- Si les sale algo como `git version 2.43.0`, ya lo tienen instalado. Salten a la sección 8.4 (configuración).
- Si les sale `command not found` o `git no se reconoce como comando`, sigan leyendo.


### 8.2 Instalación en **Mac**

La forma más sencilla es aprovechar que macOS trae las *Command Line Tools* de Xcode. En Terminal:

```bash
xcode-select --install
```

Se abre un diálogo gráfico pidiendo aceptar la instalación → acepten. Tarda unos minutos.

**Alternativa moderna (recomendada si van a hacer desarrollo seriamente)**: instalar [Homebrew](https://brew.sh) y usar `brew install git`:

```bash
# 1. Instalar Homebrew (si no lo tienen)
/bin/bash -c "$(curl -fsSL https://raw.githubusercontent.com/Homebrew/install/HEAD/install.sh)"

# 2. Instalar git
brew install git
```

Homebrew es un "App Store" para herramientas de línea de comandos. Van a usarlo mucho.

Una vez termine, cierren y reabran Terminal, y verifiquen:

```bash
git --version
```


### 8.3 Instalación en **Windows**

En Windows la forma oficial es descargar **Git for Windows**, que incluye Git + una terminal buena (Git Bash) + una GUI simple:

1. Vayan a [git-scm.com/download/win](https://git-scm.com/download/win) — la descarga arranca automáticamente.
2. Ejecuten el instalador `.exe`.
3. Durante el instalador, acepten las opciones por defecto **excepto**:
   - En "Choosing the default editor used by Git" → si no son usuarios de Vim, elijan **Visual Studio Code** o **Notepad**. Vim es una pesadilla para principiantes.
   - En "Adjusting your PATH environment" → dejen marcado **"Git from the command line and also from 3rd-party software"** (opción del medio, es la default).
4. Terminen la instalación.

Después, en PowerShell:

```bash
git --version
```

Si el comando no se reconoce, cierren y reabran PowerShell. Si aún así no funciona, **reinicien** el computador y vuelvan a probar.

> **Tip**: en Windows muchos prefieren usar **Git Bash** (que viene con Git for Windows) en vez de PowerShell. Git Bash les da una terminal estilo Linux, que es lo que verán en todos los tutoriales de internet.


### 8.4 Configurar Git (ambos sistemas operativos)

Una vez instalado, Git necesita saber **quién** son ustedes para marcar cada commit. Esto se hace una sola vez por computador.

En Terminal / PowerShell / Git Bash:

```bash
# Reemplacen con SU nombre y SU correo de GitHub
git config --global user.name "Lucas Gomez"
git config --global user.email "lucasgomeztobon@gmail.com"

# Recomendado: que la rama por defecto se llame 'main' (no 'master')
git config --global init.defaultBranch main

# Recomendado: mejor herramienta de diff en la terminal
git config --global pull.rebase false
```

Verifiquen que quedó bien:

```bash
git config --global --list
```

Debería aparecer `user.name=...` y `user.email=...`.

> **Importante**: el correo debe ser **el mismo** que usan en GitHub para que GitHub pueda asociar sus commits a su perfil.

````python
# Celda de verificacion 3 -- los datasets estan presentes y se leen?
from pathlib import Path
import pandas as pd

# Resolver el root del repo: si este notebook se abre desde notebooks/,
# cwd puede ser notebooks/ o el root. Buscamos la carpeta data/ subiendo si hace falta.
def repo_root():
    cwd = Path.cwd()
    for candidate in [cwd, cwd.parent]:
        if (candidate / "data").is_dir():
            return candidate
    raise FileNotFoundError("No se encontro la carpeta data/. Abre el notebook desde el repo.")

ROOT = repo_root()
DATA = ROOT / "data"
print("Repo root:", ROOT)

adm_path = DATA / "admissions.csv"
bmi_path = DATA / "bmi.csv"

assert adm_path.exists(), f"No se encuentra {adm_path}"
assert bmi_path.exists(), f"No se encuentra {bmi_path}"

adm = pd.read_csv(adm_path)
bmi = pd.read_csv(bmi_path, sep=";")

print(f"admissions.csv  -> {adm.shape[0]} filas, {adm.shape[1]} columnas")
print(f"bmi.csv         -> {bmi.shape[0]} filas, {bmi.shape[1]} columnas")
print()
print("Columnas admisiones:", list(adm.columns))
print("Columnas BMI:       ", list(bmi.columns))
````

````python
# Celda de verificacion 4 -- cuantos matches hay entre los datasets?
# (esto lo explora la proxima clase con prompt engineering)
from pathlib import Path
import pandas as pd

def repo_root():
    cwd = Path.cwd()
    for candidate in [cwd, cwd.parent]:
        if (candidate / "data").is_dir():
            return candidate
    raise FileNotFoundError("No se encontro la carpeta data/.")

DATA = repo_root() / "data"
adm = pd.read_csv(DATA / "admissions.csv")
bmi = pd.read_csv(DATA / "bmi.csv", sep=";")

ids_adm = set(adm["Student_ID"])
ids_bmi = set(bmi["Student_ID"])
comunes = ids_adm & ids_bmi

print(f"Student_IDs en admisiones: {len(ids_adm)}")
print(f"Student_IDs en BMI:        {len(ids_bmi)}")
print(f"Student_IDs en AMBOS:      {len(comunes)}")
print(f"Admisiones SIN match BMI:  {len(ids_adm - ids_bmi)}")
print(f"BMI SIN admision:          {len(ids_bmi - ids_adm)}")
````
